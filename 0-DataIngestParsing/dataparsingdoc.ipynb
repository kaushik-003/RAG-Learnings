### Word Doc Processing


```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
# using Docx2txtLoader
print ("Doc2textLoader")

try:
    docx_loader = Docx2txtLoader("data/docs/CHAPTER 5 _ INDIA , THAT IS BHARAT.docx")
    docs = docx_loader.load()
    print(f"  Loaded {len(docs)} document(s) using Docx2txtLoader")
    print(f"  content preview: {docs[0].page_content[:200]}...")
    print(f"  Metadata: {docs[0].metadata}")

except Exception as e:
    print(f"Error : {e}")

```

```python
# using UnstructuredWordDocumentLoader

print ("\nUnstructuredWordDocumentLoader")

try:
    unstructured_loader = UnstructuredWordDocumentLoader("data/docs/CHAPTER 5 _ INDIA , THAT IS BHARAT.docx", mode = "elements")
    unstructured_docs = unstructured_loader.load()
    print(f" Loaded {len(unstructured_docs)} elements")

    for i, doc in enumerate(unstructured_docs[:3]):
        print(f"Element {i+1} :")
        print(f"Type: {doc.metadata.get('category', 'unknown')}")
        print(f"Content preview: {doc.page_content[:100]}...\n")

except Exception as e:
    print(f"Error : {e}")
    
```
